```
# Task 3: Calculate the inverse of a sqaure matrix using np.svd
# also verify the results 
X = np.random.random((4,4))
u, d, vh = np.linalg.svd(X)
inv = vh.T @ np.diag(1/d) @ u.T
print(inv)
print(np.linalg.inv(X))
#Check
print(np.allclose(inv@X,np.eye(4)))
```

**Question:**

Can you derive the formula to compute distance between two parallel lines from the formulas provided in the previous section?

```
from math import sqrt
def distance_from_origin(a,b,c):
  return abs(c) / sqrt(a*a+b*b)  
```

```
distance_from_origin(1,0,-50) # x = 50 line
```

```
distance_from_origin(0,1,-25) # y = 25 line
```

```
def distance_between_lines(a,b,c1, c2):
  return abs(c1-c2) / sqrt(a*a+b*b)
```

```
distance_between_lines(2,1,9,5)
```

```
from mpl_toolkits.mplot3d import Axes3D # Importing this is necessary when you plot in 3D
```

```
a = 1
b = 1
c = 0
d = 10
xx = np.linspace(-10, 10, 20)
zz = np.linspace(-10, 10, 20)
X, Z = np.meshgrid(xx, zz)
yy = (-a*X - c*Z - d) * 1.0 / b
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X, yy, Z)
```
